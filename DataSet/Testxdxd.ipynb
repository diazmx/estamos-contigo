```python
import numpy as np
import pandas as pd
```

```python
tweets = pd.read_csv("tweets.csv")
```

```python
new_d = tweets["Tweet Text"]
new_d[]
```

```python
listSen = []
for t in new_d:
    text = tb(t)
    try:
        t = text.translate(to="en")
        listSen.append(t.sentiment.polarity)
    except:
        print("El elemento no est√° presente")
```

```python
pdSen = pd.DataFrame(listSen)
pdSen[pdSen[0] < 0]
```

```python
new_d = pd.DataFrame(new_d)
```
